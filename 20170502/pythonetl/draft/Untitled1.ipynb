```python
import re

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
res = requests.post("http://coolpc.com.tw/m/m-list.php", data={"G":6})
```

```python
res
```

```python
soup = BeautifulSoup(res.text,'lxml')
```

```python
mem = soup.select("span.Q")[0].find_all(attrs={"onclick": 'A(this)'})
mem
```

```python
data = [row.text for row in mem]
data
```

```python
data[1]
```

```python
re.findall('(\d+)G',data[1])
```

```python
re.findall('-(\d+)',data[1])
```

```python
re.findall('DDR(\d)',data[1])
```

```python
re.findall('\$(\d+)',data[1])
```

```python
result = []
for row in data:
    mem = {}
    try:
        mem['name'] = (row.split("/")[0])[:25]
        mem['capacity']=int(re.findall('(\d+)G', row)[0])
        mem['frequency']=int(re.findall('-(\d+)',row)[0])
        mem['ddr']=int(re.findall('DDR(3|4)', row)[0])
        mem['price']=int(re.findall('\$(\d+)', row)[0])
        result.append(mem)
        
#         ssd['read'] = int(re.findall('讀:(\d+)M', row)[0])
#         ssd['write'] = int(re.findall('寫:(\d+)M', row)[0])
#         ssd['price'] = int(re.findall('\$(\d+)', row)[0])
#         ssd['warranty'] = re.findall('(.)年', row)[0]
#         result.append(ssd)
    except:
        print(row)
result
```

```python
df=pd.DataFrame(result)
```

```python
df.head()
```

```python
df['cp']=df.price /df.capacity
df.head()
```
