````python
with open('./datasets/trump_speech.txt') as f:
    speech = f.read()
````

````python
with open('./datasets/eng_stop_words.txt') as f:
    stop_words = f.read().split('\n')
````

````python
stop_words
````

````python
speech[:100]
````

````python
from collections import Counter
wc = Counter() # Dictionary like object

for word_raw in speech.split():
    word = word_raw.replace(',', '').replace('.', '').lower()
    if word not in stop_words:
        if word in wc:
            wc[word] += 1
        else:
            wc[word] = 1
````

````python
# [{'1': 'record 1'},
#  {'2': 'record 2'}]
gym_data = []
with open('./datasets/gym.csv') as f:
    for idx, row in enumerate(f): # enumerate -> 得到for loop的index
        if idx == 0:
            header = row.strip().split(",")
        else:
            # zip(['key1', 'key2], [1,2]) -> (key1, 1), (key2, 2)
            # dictionary comprehension..
            row_dict = {k:v for k,v in zip(header, row.strip().split(","))}
            gym_data.append(row_dict)
        print("NO %s - %s" % (idx, row.strip()))
````

## 自動換頁爬蟲

````python
import requests
from bs4 import BeautifulSoup
````

````python
HOST = "https://www.ptt.cc"

headers = {"cookie": "over18=1;"}
````

````python
res = requests.get(HOST + "/bbs/NBA/index.html", headers=headers)
soup = BeautifulSoup(res.text, 'lxml')
````

````python
# 爬取換頁資訊
buttons = soup.select('a.btn.wide')
buttons
````

````python
# 取得第一個
buttons[1]
````

````python
buttons[1]['href']
````

````python
total_page = int(buttons[1]['href'].split('index')[1].split('.html')[0]) + 1
total_page
````

````python
page_to_crawl = 5
for page in range(total_page, total_page - page_to_crawl, -1):
    print(page)
````

````python
# 有了頁面資訊，我們來寫可以換頁的列表爬蟲
page_to_crawl = 5
for page in range(total_page, total_page - page_to_crawl, -1):
    print("[DEBUG] Current page {}".format(HOST + "/bbs/NBA/index{}.html".format(page)))
    res = requests.get(HOST + "/bbs/NBA/index{}.html".format(page), headers=headers)
    soup = BeautifulSoup(res.text, 'lxml')
    links = soup.select('div.title > a')
    for link in links:
        print("link is %s" % HOST + link['href'])
        print("title is %s" % link.text)
        print("=" * 50)
````
