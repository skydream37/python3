```python
data = []
with open('./gym.csv') as f:
    for no, line in enumerate(f):
        if no == 0:
            header = line.strip().split(",")
            print("[DEBUG] header is %s" % header)
        else:
            row_dict = {k: v for k,v in zip(header, line.strip().split(","))}
            data.append(row_dict)
```

```python
for idx, row in enumerate(data):
    if not row['pool']:
        print("bad line, %s" % row)
        del data[idx]
```

```python
time_serie = [dateutil.parser.parse(x['timestamp']) for x in data]
```

```python
pool = [int(x['pool']) for x in data]
```
