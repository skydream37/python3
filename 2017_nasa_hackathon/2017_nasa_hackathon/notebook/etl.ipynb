```python
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
from datetime import timedelta, timezone
```

```python
tpe = timezone(+(timedelta(hours=8)))
```

```python
parser = lambda x: pd.to_datetime(x, format="%Y%m%d", errors='coerce')
```

```python
df = pd.read_csv('./water.csv', parse_dates=['date'], date_parser=parser)
```

```python
df.shape
```

```python
df2 = df[df.date.notnull()]
```

```python
df2.head()
```

```python
df2['timestamp'] = df2.date.apply(lambda x: x.replace(tzinfo=tpe).timestamp())
```

```python
df2.head()
```

```python
del df2['id']
```

```python
del df2['date']
```

```python
df2.head()
```

```python
well_meta = pd.read_csv('./well_meta.csv')
```

```python
well_meta
```

```python
df = pd.merge(df2, well_meta, how='inner', on='wellid')
```

```python
df.head()
```

```python
scaler = MinMaxScaler()
```

```python
df3 = pd.DataFrame()
```

```python
for row in df.wellid.unique():
    ddf = df[df.wellid == row].copy()
    ddf['l'] = scaler.fit_transform(df[df.wellid == row]['level'])
    df3 = df3.append(ddf)
```

```python
df3
```

```python
del df3['level']
```

```python
df3['level'] = df3['l']
```

```python
del df3['l']
```

```python
df3.head()
```

```python
del df3['wellname']
```

```python
del df3['wellid']
```

```python
df3
```

```python
df3['type'] = 'well'
```

```python
df3.head()
```

```python
df3.to_csv('./well_data.csv')
```
