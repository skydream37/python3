```python
df2 = df[(df['time'] > 1388505600000) & (df['time'] < 1420041600000)]
```

```python
del df2['timestamp']
```

```python
def df_to_geojson(df, properties, lat='lat', lon='lon'):
    geojson = {'type':'FeatureCollection', 'features':[]}
    for _, row in df.iterrows():
        feature = {'type':'Feature',
                   'properties':{},
                   'geometry':{'type':'Point',
                               'coordinates':[]}}
        feature['geometry']['coordinates'] = [row[lon],row[lat]]
        for prop in properties:
            feature['properties'][prop] = row[prop]
        if row['level'] > 0.7:
            feature['properties']['style'] = {"fillColor": "blue"}
        elif row['level'] > 0.3 and row['level'] < 0.7:
            feature['properties']['style'] = {"fillColor": "yellow"}
        else:
            feature['properties']['style'] = {"fillColor": "red"}
        geojson['features'].append(feature)
    return geojson
```

```python
gjson = df_to_geojson(df2, ['level', 'time'])
```
